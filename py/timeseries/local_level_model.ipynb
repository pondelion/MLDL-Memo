```
import numpy as np
import matplotlib.pyplot as plt
```

```
def local_level_model(
    x_prev: float,
    W: float,
    V: float,
):
    G = 1
    F = 1
    x = G*x_prev + np.random.normal(loc=0, scale=W, size=1)
    y = F*x + np.random.normal(loc=0, scale=V, size=1)

    return x, y
```

```
W = 1  # 状態雑音標準偏差
V = 2  # 観測雑音標準偏差
```

```
x_hist = [0]
y_hist = []
```

```
for i in range(1000):
    x, y = local_level_model(
        x_hist[-1],
        W, V
    )
    x_hist.append(x)
    y_hist.append(y)
```

```
plt.figure(figsize=(12, 6))
plt.plot(x_hist, alpha=0.6, label='state')
plt.plot(y_hist, alpha=0.6, label='observation')
```
