```python
import random
import os
import shutil
```

```python
def sample_wav(n):
    wavs = []
    texts = []
    with open('/content/audio/text.txt', 'r') as f:
        lines = f.readlines()
        lines_sampled = random.sample(lines, n)
        for line in lines_sampled:
            wav_path, text = line.split(':')
            wav_path = os.path.join('/content/audio/wav', os.path.basename(wav_path))
            wavs.append(wav_path)
            texts.append(text.replace('\n', ''))
    return wavs, texts


def all_wav():
    wavs = []
    texts = []
    with open('/content/audio/text.txt', 'r') as f:
        lines = f.readlines()
        lines_sampled = lines #random.sample(lines, n)
        for line in lines_sampled:
            wav_path, text = line.split(':')
            wav_path = os.path.join('/content/audio/wav', os.path.basename(wav_path))
            wavs.append(wav_path)
            texts.append(text.replace('\n', ''))
    return wavs, texts
```

```python
import librosa
import soundfile as sf
```

```python
def copy_wav(wavs, dest_dir):
    os.makedirs(dest_dir, exist_ok=True)
    for wav in wavs:
        wav_name = f'SPK1_{os.path.basename(wav)}'
        y, sr = librosa.core.load(wav, sr=44100, mono=True)
        sf.write(os.path.join(dest_dir, wav_name), y, sr, subtype="PCM_16")
        # shutil.copyfile(wav, os.path.join(dest_dir, wav_name))
```

```python
def create_transcript_textfile(wavs, texts, filepath='/content/espnet/egs2/custom_dataset/tts1/dataset/transcript.txt'):
    with open(filepath, 'w') as wf:
        for wav, text in zip(wavs, texts):
            wav_name = f'SPK1_{os.path.basename(wav)}'
            wav_name = wav_name.replace('.wav', '')
            text = text.replace("\n", "")
            wf.write(f'{wav_name}:{text}' + '\n')
```
