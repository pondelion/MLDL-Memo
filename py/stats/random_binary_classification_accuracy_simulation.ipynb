```
import numpy as np
from fastprogress import progress_bar as pb
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

### 二値分類でランダムに分類した場合のAccuracy0.5から逸脱する確率を各データ数でシミュレーション

```
data_nums = [10, 100, 1000, 10000, 10000]
N_TRY = 10000
acc_dict = {}

for n_data in pb(data_nums):
    accs = []
    for _ in range(N_TRY):
        true_labels = np.random.randint(0, 2, n_data)
        pred_labels = np.random.randint(0, 2, n_data)
        acc = (true_labels==pred_labels).sum() / n_data
        accs.append(acc)
    acc_dict[n_data] = accs
```

```

for i, (n_data, accs) in enumerate(acc_dict.items()):
    plt.hist(accs, bins=int(40/(i*3+1)))
    plt.title(f'sample data num : {n_data}')
    plt.xlabel('Accuracy')
    plt.xlim(-0.1, 1.1)
    print(f'データ数{n_data}では、Accuracyが{np.percentile(accs, 95)}以上になる確率は5%、{np.percentile(accs, 5)}以下になる確率は5%')
    plt.show()
```
