```
import torch
from torchvision import models, transforms
```

```
resnet18 = models.resnet18(pretrained=True)
alexnet = models.alexnet(pretrained=True)
squeezenet = models.squeezenet1_0(pretrained=True)
vgg16 = models.vgg16(pretrained=True)
```

```
densenet = models.densenet161(pretrained=True)
```

```
inception = models.inception_v3(pretrained=True)
```

```
googlenet = models.googlenet(pretrained=True)
```

```
shufflenet = models.shufflenet_v2_x1_0(pretrained=True)
mobilenet = models.mobilenet_v2(pretrained=True)
resnext50_32x4d = models.resnext50_32x4d(pretrained=True)
wide_resnet50_2 = models.wide_resnet50_2(pretrained=True)
mnasnet = models.mnasnet1_0(pretrained=True)
```

```
torch.save(resnet18.state_dict(), './resnet18.model')
torch.save(alexnet.state_dict(), './alexnet.model')
torch.save(squeezenet.state_dict(), './squeezenet.model')
torch.save(vgg16.state_dict(), './vgg16.model')
torch.save(densenet.state_dict(), './densenet.model')
torch.save(googlenet.state_dict(), './googlenet.model')
torch.save(shufflenet.state_dict(), './shufflenet.model')
torch.save(mobilenet.state_dict(), './mobilenet.model')
torch.save(resnext50_32x4d.state_dict(), './resnext50_32x4d.model')
torch.save(wide_resnet50_2.state_dict(), './wide_resnet50_2.model')
torch.save(mnasnet.state_dict(), './mnasnet.model')
torch.save(inception.state_dict(), './inception.model')
```
